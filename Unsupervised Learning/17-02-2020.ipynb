```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,Y = make_blobs(n_samples = 300, centers = 4, 
                 cluster_std = 0.60,random_state = 0)
```

```python
plt.scatter(X[:,0],X[:,1],s = 50)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```python
plt.scatter(X[:,0],X[:,1],c = y_pred,s = 50)
```

```python
centers = kmeans.cluster_centers_
```

```python
centers
```

```python
plt.scatter(X[:,0],X[:,1],c = y_pred,s = 50)
plt.scatter(centers[:,0],centers[:,1],c = 'black',s = 200,alpha = 0.5)
```

```python
from sklearn import preprocessing
df_standard = preprocessing.scale(df)
```

```python
df_standard
```

```python
df_standard = pd.DataFrame(df_standard)
```

```python
df_standard
```

```python
from sklearn.cluster import KMeans
l1 = []
for i in range(1,11):
    kmeans2 = KMeans(n_clusters = i,init = 'k-means++',random_state = 1)
    kmeans2.fit(df_standard)
    l1.append(kmeans2.inertia_)
```

**inertia_ : Sum of squared distance to their closest cluster center**

```python
y_pred = kmeans2.fit_predict(df_standard)
```
